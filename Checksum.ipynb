```python

def complement(x):
    result=[]
    for i in range(0,len(x)):
        if x[i]=='0':
            result.append('1')
        else:
            result.append('0')
    return ''.join(result)
def bin_sum(a,b):
    n=len(a)
    intsum=int(a,2)+int(b,2)
    result=bin(intsum)[-n:]
    if a[0]=='1' and b[0]=='1':
        intsum=int(result,2)+int(('0'*(n-1)+'1'),2)
        result=bin(intsum)[-n:]
    for i in range (0,len(result)):
        if result[i]=='b':
            t=[]
            t=list(result)
            t[i]='0'
            result=''.join(t)
    return result
def sender_checksum(frames):
    print("\n\t\tSender Side",end="\n")
    x=""
    for i in range(0,len(frames)):
        x+=frames[i]
    print("Dataword=",x,end="\n")
    for i in range(0,len(frames)):
        print("Segment%d=%s"%(i+1,frames[i]),end="\n")
    n=len(frames[0])
    s='0'*(n)
    for i in range(0,len(frames)):
        s=bin_sum(s,frames[i])      
    print("SUM=",s)
    s=complement(s)
    return s

def receiver_checksum(frames,checksum):
    print("\n\t\tReceiver Side",end="\n")
    x=""
    for i in range(0,len(frames)):
        x+=frames[i]
    print("Codeword=",x+checksum,end="\n")
    n=len(frames[0])
    s='0'*(n)
    for i in range(0,len(frames)):   
        s=bin_sum(s,frames[i])
    s=bin_sum(s,checksum)
    ran=random.randint(0,2)
#     ran=0                  #for corrupting bits at receiver end
    if ran==0:
        l2=list(s)
        if l2[0]=="1":
            l2[0]="0"
        else:
            l2[0]="1"
        s=''.join(l2)
    s=complement(s)
    print("Checksum=",s)
    if int(s,2)==0:
        print("Data Accpeted")
        for i in range(0,len(frames)):
            print("Segment%d=%s"%(i+1,frames[i]),end="\n")
        x=""
        for i in range(0,len(frames)):
            x+=frames[i]
        print("Dataword Received=",x,end="\n")
    else:
        print("Data Rejected")
        
n=int(input("Enter segment size="))
l=[]
frame=input("Enter Dataword=")
m=len(frame)/n
i=0
while i<len(frame):
    l.append(frame[i:(i+n)])
    i+=n
checksum=sender_checksum(l)
print("Checksum=",checksum)
receiver_checksum(l,checksum)
```
